```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x **2
```

```python
y
```

**Matplot을 함수형/클래스형 2가지로 사용할 수 있다.**

```python
# Functional
plt.plot(x,y)
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Title")
```

**Multi plot on same canvas**

```python
plt.subplot(1,2,1)
# 1행 2열 중 첫번째
plt.plot(x,y,"r")
plt.subplot(1,2,2)
# 1행 2열 중 두번째
plt.plot(y,x,"b")
```

**OOP version**

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x,y)
axes.set_xlabel("X label")
axes.set_ylabel("Y label")
axes.set_title("Set Title")
```

**fig = empty canvas**  
**then add an axes to the figure (Manually)**

```python
fig = plt.figure()
axes0 = fig.add_axes([0,0,1,1])
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
axes1.set_title("Larger plot")
axes2.set_title("Smaller plot")
```

```python
fig,axes = plt.subplots()
axes.plot(x,y)
```

**subplots = add_axes를 자동으로 해주는 것. based on how many rows and columns you want.**

**axes = list of these axes object. iteration 가능**

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
# to fix overlapping
plt.tight_layout()
for current_ax in axes:
    current_ax.plot(x,y)
```

**indexing 가능**

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y)
axes[0].set_title("First Plot")
axes[1].plot(y,x)
axes[1].set_title("Second Plot")
# 항상 마지막에 안 겹치게 tight_layout 넣어주는게 좋다.
plt.tight_layout()
```

# Figure Size and DPI

```python
fig = plt.figure(figsize=(8,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

**Saving into image file**

```python
fig.savefig("my_picture.png", dpi=200)
```

**Adding legend**
> legend looks for label. Without label, there will be an error
> legend location with loc code or tuple

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, label="square")
ax.plot(x,x**3, label="Cubed")
# location of legend by loc code
# 0 for best, 1 for upper-right, 2 for upper-left, 3 forlower-left ... 10 for center
# ax.legend(loc=10)
# specifying using tuple
ax.legend(loc=(0.1,0.1))
```

# Customizing Plot Appearance

**Colors**

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
# basic color
# ax.plot(x,y, color="orange")
# RGB He code
ax.plot(x,y, color="#FF8C00")
```

**Default - linewidth=1, lw for shortcut**  
**Transparency: alpha=1**

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color="purple", lw=3, alpha=0.5)
```

**Line Style, shortcut: ls**
> options are
1. **--**
2. **-.**
3. **:**
4. **steps**
5. **-** for solid line

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color="purple", lw=3, linestyle="--")
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color="purple", lw=3, linestyle="-.")
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color="purple", lw=3, linestyle=":")
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color="purple", lw=3, linestyle="steps")
```

# Markers
1. o
2. +
3. *
4. 1

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color="purple", lw=0.5, linestyle="-", marker="1", markersize=20)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color="purple", lw=0.5, linestyle="-", marker="o", markersize=20, markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green")
```

# Plot Range

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color="purple", lw=2, ls="--")
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
